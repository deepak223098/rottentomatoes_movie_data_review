```python
payload = {
'direction': 'next',
'endCursor': '',
'startCursor': '',
}
```

```python
import time
import re
```

```python
import requests
s = requests.Session()
```

```python
r = s.get(url, headers=headers, params=payload) # GET Call
data = r.json()
```

```python
movie_data = {"ReviewID":None,"Reviewer Name":None,"Review":None,"Rating":None,"Date-of-Review":None}
```

```python
def movie_update_dict(data):
    for i in range(len(data['reviews'])):
        if movie_data.get('ReviewID'):
            temprid,temprname,temprev,temprating,tempdate = movie_data.get('ReviewID'),\
            movie_data.get("Reviewer Name"),movie_data.get("Review"),movie_data.get("Rating"),movie_data.get("Date-of-Review")
        else:
            temprid,temprname,temprev,temprating,tempdate = [],[],[],[],[]
        # End if
        if (data['reviews'][i]['user']['userId'].isdigit()) :#and (data['reviews'][i]['displayName'])
            temprid.append(data['reviews'][i]['user']['userId'])
            temprname.append(data['reviews'][i]['displayName'])
            temprev.append(data['reviews'][i]['review'])
            vrating = float(".".join(re.findall(r'\d+',data['reviews'][i]['rating']))) if ".".join(re.findall(r'\d+',data['reviews'][i]['rating'])) else ""
            temprating.append(vrating)
            tempdate.append(data['reviews'][i]['updateDate'])
            movie_data["ReviewID"] = temprid.copy()
            movie_data["Reviewer Name"] = temprname.copy()
            movie_data["Review"] = temprev.copy()
            movie_data["Rating"] = temprating.copy()
            movie_data["Date-of-Review"] = tempdate.copy()
        else:
            pass
    return movie_data
movie_update_dict(data)
```
